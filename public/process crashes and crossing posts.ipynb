```python
crossing_posts = gpd.read_file('./crossing_locations.geo.json')
```

```python
re.findall('([-]?\d+.\d+)','POINT (-73.92184 40.86527)')
```

```python
crossing_posts[['LONGITUDE','LATITUDE']] = crossing_posts['geometry'].apply(
    lambda x: pd.Series(re.findall(r'([-]?\d+.\d+)', str(x)))
)

crossing_posts[[
    'POST #',
    'SCHOOL NAME',
    'LAST NAME',
    'STREET NAME 1',
    'STREET NAME 2',
    'AM HOURS',
    'PM HOURS',
    'LONGITUDE',
    'LATITUDE'
]].to_csv('crossing_posts.csv', index = False)
```
